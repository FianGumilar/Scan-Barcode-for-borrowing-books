```python
from base64 import decode
from fileinput import filename
from sre_constants import SUCCESS
import cv2
import winsound
import csv
import matplotlib.pyplot as plt
import numpy as np
import datetime
#from dateutil.relativedelta import relativedelta
from pyzbar.pyzbar import decode #konversi dari format yang disandikan

cap = cv2.VideoCapture(1)
cap.set(3,640)
cap.set(4, 480)


with open('myDataFile.txt') as f:
    myDataList = f.read().splitlines()

while True:

    ret, frame = cap.read()

    if not ret:
        break

    imgbook = cap
    dt = datetime.datetime.now()
    dtime = dt.strftime('%d/%m/%Y %H:%M')
    #date_end = datetime.datetime.now()+ relativedelta(days=14)
    #dreturn = date_end.strftime('%d/%m/%Y')
    img = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    gray = cv2.cvtColor(frame, 0)

    colored_negative = abs(255-img)
    gray_negative = abs(255-gray)

    imgs = [img, gray, colored_negative, gray_negative]
    title = ['coloured', 'gray', 'coloured-negative', 'gray-negative']

    plt.subplot(2, 2, 1)
    plt.title(title[0])
    plt.imshow(imgs[0])
    plt.xticks([])
    plt.yticks([])

    plt.subplot(2, 2, 2)
    plt.title(title[1])
    plt.imshow(imgs[1], cmap='gray')
    plt.xticks([])
    plt.yticks([])

    plt.subplot(2, 2, 3)
    plt.title(title[2])
    plt.imshow(imgs[2])
    plt.xticks([])
    plt.yticks([])

    plt.subplot(2, 2, 4)
    plt.title(title[3])
    plt.imshow(imgs[3], cmap='gray')
    plt.xticks([])
    plt.yticks([])

    for cartbarcode in decode(gray_negative):
        myData1 = cartbarcode.data.decode('utf-8')
        winsound.PlaySound('beepsound.wav', winsound.SND_ASYNC)

    for bookbarcode in decode(gray):
        myData2 = bookbarcode.data.decode('utf-8')
        winsound.PlaySound('beepsound.wav', winsound.SND_ASYNC)

        if myData1 in myDataList:
            print(myData1,'Allowed',myData2, dtime)
            winsound.PlaySound('returnbook.wav', winsound.SND_ASYNC)

        else:
            print('Un-Allowed')
            winsound.PlaySound('no-success.wav', winsound.SND_ASYNC)

        ptscart = np.array([cartbarcode.polygon],np.int32)
        ptscart = ptscart.reshape((-1,1,2))
        ptsbook = np.array([bookbarcode.polygon],np.int32)
        ptsbook = ptscart.reshape((-1,1,2))
        cv2.polylines(gray_negative,[ptscart],True,(255,0,255),5)
        pts2 = cartbarcode.rect
        pts3 = bookbarcode.rect
        cv2.putText(gray_negative,myData1,(pts2[0],pts2[1]),cv2.FONT_HERSHEY_SIMPLEX, 0.9,(255,0,255),2)    
        cv2.putText(img,myData1,(pts3[0],pts3[1]),cv2.FONT_HERSHEY_SIMPLEX, 0.9,(255,0,255),2)   

        d = open('data-pengembalian.csv', 'a', newline="")
        data = (myData1, 'Allowed', myData2, dtime)
        writer = csv.writer(d)
        writer.writerow(data)
        f.close()

    cv2.imshow('CartBarcode', gray_negative) 
    cv2.imshow('BookBarcode', img)


    if cv2.waitKey(1) == 27:
            break
  

cap.release()
cv2.destroyAllWindows()
```
